```python
import matplotlib.pyplot as plt
# Convert genre_ids to category 
df4['genre_ids'] = df4['genre_ids'].astype('category')
# Aggregate data
genre_gross = df4.groupby('genre_ids')[['domestic_gross', 'foreign_gross']].sum()

# ... (your data aggregation code for genre_gross) ...

# Create bar chart
plt.figure(figsize=(12, 6))  # Correct way to create a figure
plt.bar(genre_gross.index, genre_gross['domestic_gross'], label='Domestic Gross')
plt.bar(genre_gross.index, genre_gross['foreign_gross'], bottom=genre_gross['domestic_gross'], label='Foreign Gross') 
plt.title('Domestic and Foreign Gross by Genre')
plt.ylabel('Gross Revenue')
plt.xlabel('Genre')
plt.xticks(rotation=45)
plt.legend()
plt.show()

```

```python
print(df4[['domestic_gross', 'foreign_gross']].dtypes)
```

```python
# convert to numeric
df4['domestic_gross'] = pd.to_numeric(df4['domestic_gross'], errors='coerce')
df4['foreign_gross'] = pd.to_numeric(df4['foreign_gross'], errors='coerce')
```

```python
# 1. Drop rows with NaN values (resulting from the conversion)
df4 = df4.dropna(subset=['domestic_gross', 'foreign_gross'])
# 2. Group by year and calculate mean gross
yearly_avg_gross = df4.groupby('year')[['domestic_gross', 'foreign_gross']].mean()
```

```python


# Create line chart
plt.figure(figsize=(10, 5))
plt.plot(yearly_avg_gross.index, yearly_avg_gross['domestic_gross'], label='Domestic Gross')
plt.plot(yearly_avg_gross.index, yearly_avg_gross['foreign_gross'], label='Foreign Gross')
plt.title('Average Domestic and Foreign Gross Over Time')
plt.xlabel('Year')
plt.ylabel('Average Gross Revenue')
plt.legend()
plt.grid(axis='y', linestyle='--')  # Add a subtle grid for easier reading
plt.show()
```

```python
# Get top 5 studios
top_studios = df4.groupby('studio')['domestic_gross'].sum().nlargest(5)

# Create bar chart
plt.figure(figsize=(8, 4))
top_studios.plot(kind='bar')
plt.title('Top 5 Studios by Domestic Gross')
plt.ylabel('Domestic Gross')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df4['popularity'], df4['domestic_gross'], alpha=0.5)
plt.title('Domestic Gross vs. Popularity')
plt.xlabel('Popularity')
plt.ylabel('Domestic Gross')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns  # Seaborn for enhanced plotting

# Aggregate data by genre
genre_agg = df4.groupby('genre_ids')[['domestic_gross', 'foreign_gross', 'popularity']].mean().reset_index()

# Create scatter plot using Seaborn
plt.figure(figsize=(10, 8))
sns.scatterplot(data=genre_agg, x='popularity', y='domestic_gross', hue='genre_ids', palette='viridis', s=100)
plt.title('Domestic Gross vs. Popularity by Genre')
plt.xlabel('Popularity')
plt.ylabel('Domestic Gross')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.show()
```
